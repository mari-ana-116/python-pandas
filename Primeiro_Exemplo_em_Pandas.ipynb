```python
# Quais são as colunas e que tipo de dados cada um tem com .info
print(combustiveis_df.info())
```

```python
# describe() mostra as estatísticas mais básicas que a gente precisa
print(combustiveis_df.describe())
```

```python
# Filtrar apenas por uma coluna
display(combustiveis_df['Revenda'])
```

```python
# Aqui criamos um novo dataframe apenas com as colunas que eu quero
ca_df = combustiveis_df[['Revenda', 'Municipio', 'Produto', 'Valor de Venda']]
display(ca_df)
```

```python
# Exibe a 4ª linha
display(ca_df.loc[3])
```

```python
# Exibir da 10ª linha até a 20ª linha
display(ca_df.loc[9:19])
```

```python
# Criar um dataframe gas_df contendo 
# apenas as 4 colunas (Revenda, Municipio, Produto e Valor de Venda)
# somente com combustível sendo GASOLINA e exibir na tela
gas_df = ca_df.loc[ca_df['Produto'] == 'GASOLINA']
display(gas_df)

# Exibe o valor de gasolina mais caro com .max()

display(gas_df['Valor de Venda'].max())
```

```python
# Se eu quiser saber qual é o posto e município da gasolina mais cara do país, basta utilizar .max()
display(gas_df[['Revenda','Municipio', 'Valor de Venda']].max())
```

```python
# DataFrame.loc[] com múltiplas condições para filtragem
# Quais são os preços e postos que vendem ETANOL na cidade de INDAIATUBA?
# Ordenado do menor para o maior preço
etanol_indaiatuba_df = ca_df.loc[(ca_df['Produto'] == 'ETANOL') & (ca_df['Municipio'] == 'INDAIATUBA')]
etanol_indaiatuba_df.sort_values(by='Valor de Venda', inplace=True)
display(etanol_indaiatuba_df)
```

```python
# Qual a média de preços dos combustíveis GASOLINA e GASOLINA ADITIVADA do bairro MOOCA em SÃO PAULO
display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA') & 
                            (combustiveis_df['Municipio'] == 'SAO PAULO') & 
                            ((combustiveis_df['Produto'] == 'GASOLINA') | (combustiveis_df['Produto'] == "GASOLINA ADITIVADA")), 
                            ['Valor de Venda']].mean())

# % no sql pesquisa a palavra desejada e mostra tudo que contém aquela palavra
# | = OU, mas temos que colocar essa operação entre parênteses para que ela seeja feita primeiro e respeite os outros filtros
# Em forma de array ficaria mais fácil: (combustiveis_df['Produto'].isin(['GASOLINA', 'GASOLINA ADITIVADA'])), utilizando o .isin

# Para a média de preços, feita por último, usa-se ", ['Valor de Venda']" para obter os valores e o método ".mean()" para descobrir a média aritmética simples desses valores
```

```python
# Exemplo de uso de array utilizando .isin
display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA') & 
                            (combustiveis_df['Municipio'] == 'SAO PAULO') & 
                            (combustiveis_df['Produto'].isin(['GASOLINA', 'GASOLINA ADITIVADA'])), 
                            ['Valor de Venda']].mean())
```

```python
# Como mostrar média de valor de venda POR COMBUSTÍVEL no Brasil inteiro?
# Trabalharemos com agrupamento; grupos de DIESEL, GASOLINA, ETANOL, etc. através do .groupby
media_por_combustivel_df = ca_df[['Produto', 'Valor de Venda']].groupby(by='Produto').mean().round(2)
display(media_por_combustivel_df)

# .round(2) arredonda os valores obtidos em duas casas
```
